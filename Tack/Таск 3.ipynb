## Задание № 1
В Тридевятом Царстве уже много лет как установилась конституционная монархия. В парламент Тридевятого царства входят 26 партий, обозначаемых строчными буквами английского алфавита
от «a» до «z». Заседание парламента в соответствии с регламентом проходит по следующей схеме:
В начале заседания никаких законопроектов на повестке дня не стоит.
Любая партия (в том числе и та, которая внесла текущий законопроект) в любой момент времени может внести на обсуждение новый законопроект.
При вносе партией нового законопроекта обсуждение предыдущего приостанавливается и начинается обсуждение нового. После того, как парламентарии голосуют за новый законопроект, обсуждение
законопроекта, прерванного им, продолжается.
Дискуссия по любому законопроекту обязана завершиться голосованием по этому законопроекту.
В конце заседания не должно остаться внесённых законопроектов, дискуссия по которым не завершена.
Вам дана запись о событиях на сессии парламента. Каждое событие является либо внесением нового законопроекта, либо голосованием за какой-то законопроект, причём в обоих случаях известен номер
партии, этот законопроект предложившей. События даны в том порядке, в котором они происходили. Проверьте, может ли данная запись о событиях соответствовать какому-либо заседанию, удовлетворяющему
порядку проведения, описанному выше.

```python
import random
import string

def getChar():
    return (random.choice(string.ascii_lowercase))

def getAction():
    return (random.choice(['Add ', 'Vote ']))

def generateHistory(stop, history = []):
    party = getChar()
    if len(history) > 0:
        ind = random.randint(0, len(history))
        history.insert(ind, 'Vote ' + party)
        history.insert(ind, 'Add ' + party)
    else:
        history.insert(0, 'Vote ' + party)
        history.insert(0, 'Add ' + party)
    if len(history) == stop:
        return history
    else:
        return generateHistory(stop, history)
    
def checkHistory(history:list):
    lastAction = None
    lastParty = None
    toPop = None
    
    if len(history) % 2 == 1:
        return 'No'
    
    (action, party) = history[0].split()
    if action == 'Vote':
        return 'No'
    
    for i in range(1 + len(history) >> 1):
        
        for j in range(len(history)):
            (action, party) = history[j].split()
            
            if lastAction != None:
                if lastAction == 'Add' and action == "Vote":
                    if lastParty != party:
                        return 'No'
                    else:
                        toPop = j
                elif len(history) == 2 and lastAction == action:
                    return 'No'
                
            lastAction = action
            lastParty = party
        
        history.pop(toPop)
        history.pop(toPop-1)
        print(i, "-ая итерация:", history)
    
    return 'Yes'

k = (int)(input())

if k % 2 == 1:
    history = [getAction() + getChar()]
else:
    history = []
    
history = generateHistory(k, history)

print(history)
print(checkHistory(history), end='\n\n')

print(['Add c', 'Vote c', 'Add t', 'Add l', 'Vote l', 'Add o'])
print(checkHistory(['Add c', 'Vote c', 'Add t', 'Add l', 'Vote l', 'Add o']))
```

## Задание №2
Задан правильный N-угольник. Требуется выбрать наименьшее количество его вершин, которые также образуют правильный многоугольник.

```python
def minVertex(k:int):
    for i in range(3, (k + 1) >> 1):
        if k % i == 0:
            return i
    return k

k = (int)(input())
print(minVertex(k))
print("\n===========================\n")
k = (int)(input())
print(minVertex(k))
```

## Задание №3
На форуме, на котором обсуждаются задачи олимпиад по информатике, ввели следующий аналог капчи. Участнику выдаётся строка из N десятичных цифр (без ведущих нулей).
В качестве ответа требуется ввести такое основание системы счисления B, что в этой системе счисления выданная запись будет соответствовать составному числу (назовем его D),
а также число X, большее 1 и меньшее D, являющееся делителем D. При этом B и X не должны превосходить 10^9. По заданной строке десятичных цифр найдите любую пару чисел 
B и X, удовлетворяющую ограничениям, или ответьте, что решения в заданных ограничениях не существует.

```python
import random

def toNumsArray(num):
    numsArr = []
    while num > 0:
        numsArr.insert(0, num % 10)
        num = num // 10
    return numsArr

def minimalDivider(num):
    oldNum = num
    for i in range(2, 10):
        while num % i == 0 and num != i:
            num = num // i
    if num != oldNum:
        return num
    return -1
    
def convertToDecimal(numsArr, frm):
    convertedNum = 0
    for i in range(len(numsArr)):
        convertedNum += numsArr[i]*(frm**(len(numsArr)-i-1))
    return convertedNum

def captcha(num):
    numsArr = toNumsArray(num)
    maxNum = max(numsArr)
    
    if len(numsArr) == 1:
        minimalDiv = minimalDivider(num)
        if (minimalDiv == -1):
            return -1
        else:
            return [10, minimalDiv]
    
    for i in range(maxNum+1, 10**9):
        convertedNum = convertToDecimal(numsArr, i)
        minimalDiv = minimalDivider(convertedNum)
        if (minimalDiv > 1):
            return [i, minimalDiv]
    return -1

num = random.randint(1, 10**20)
shortNum = random.randint(1, 10**5)

print(num)
print(captcha(num), end="\n\n")

print(shortNum)
print(captcha(shortNum), end="\n\n")

print(5)
print(captcha(5), end="\n\n")


print(4)
print(captcha(4), end="\n\n")
```
